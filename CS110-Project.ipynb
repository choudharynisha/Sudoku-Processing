```java
int[][] p = new int[9][9]; // 2-D array for puzzle

int row; // the selected row
int col; // the selected column
int num; // the number that the user enters

boolean checkBox(int number) {
    int sum = 0;
    
    if(number == 0) {
        for(int i = 0; i < 3; i++) {
            for(int j = 0; j < 3; j++) {
                sum += p[i][j];
            }
        }
    } else if(number == 1) {
        for(int i = 0; i < 3; i++) {
            for(int j = 3; j < 6; j++) {
                sum += p[i][j];
            }
        }
    } else if(number == 2) {
        for(int i = 0; i < 3; i++) {
            for(int j = 6; j < 9; j++) {
                sum += p[i][j];
            }
        }
    } else if(number == 3) {
        for(int i = 3; i < 6; i++) {
            for(int j = 0; j < 3; j++) {
                sum += p[i][j];
            }
        }
    } else if(number == 4) {
        for(int i = 3; i < 6; i++) {
            for(int j = 3; j < 6; j++) {
                sum += p[i][j];
            }
        }
    } else if(number == 5) {
        for(int i = 3; i < 6; i++) {
            for(int j = 6; j < 9; j++) {
                sum += p[i][j];
            }
        }
    } else if(number == 6) {
        for(int i = 6; i < 9; i++) {
            for(int j = 0; j < 3; j++) {
                sum += p[i][j];
            }
        }
    } else if(number == 7) {
        for(int i = 6; i < 9; i++) {
            for(int j = 3; j < 6; j++) {
                sum += p[i][j];
            }
        }
    } else if(number == 8) {
        for(int i = 6; i < 9; i++) {
            for(int j = 6; j < 9; j++) {
                sum += p[i][j];
            }
        }
    }
    
    if(sum == 45) {
        return true;
    } else {
        return false;
    }
}

boolean checkColumn(int number) {
    int sum = 0;
    
    for(int i = 0; i < p.length; i++) {
        sum += p[i][num];
    }
    
    if(sum == 45) {
        return true;
    } else {
        return false;
    }
}

boolean checkCompletion() {
    for(int i = 0; i < p.length; i++) {
        for(int j = 0; j < p[i].length; j++) {
            if(p[i][j] == 0) {
                i = p.length;
                j = p[i].length;

                return false;
            }
        }
    }
    
    return true;
}

boolean checkRow(int number) {
    int sum = 0;
    
    for(int i = 0; i < p[number].length; i++) {
        sum += p[number][i];
    }
    
    if(sum == 45) {
        return true;
    } else {
        return false;
    }
}

boolean checkWin() {
    for(int i = 0; i < 9; i++) {
        if(!checkRow(i) || !checkColumn(i) || !checkBox(i)) {
            i = 9;
            return false;
        }
    }
    
    return true;
}

void choosePuzzle() {
    int puzzle = (int)(random(6) + 1);
    
    if(puzzle == 1) {
        instructions();
        puzzle1();
    } else if(puzzle == 2) {
        instructions();
        puzzle2();
    } else if(puzzle == 3) {
        instructions();
        puzzle3();
    } else if(puzzle == 4) {
        instructions();
        puzzle4();
    } else if(puzzle == 5) {
        instructions();
        puzzle5();
    } else if(puzzle == 6) {
        instructions();
        puzzle6();
    }
}

void completed() {
    if(checkWin()) {
        textSize(20);
        fill(0, 128, 0);
        text("Congratulations! You solved the puzzle", 500, 400);
        text("correctly.", 500, 430);
    } else {
        textSize(20);
        fill(255, 0, 0);
        text("Sorry! Your solution to this puzzle is", 500, 400);
        text("incorrect. Please try again!", 500, 430);
    }
}

void displayBoard() {
    textSize(30); // sets the text size to 30
    fill(0); // sets the text color to black
    
    int x = 37;
    int y = 55;
    
    for(int i = 0; i < p.length; i++) {
        for(int j = 0; j < p[i].length; j++) {
            if(p[i][j] != 0) {
                text(str(p[i][j]), x, y);
            }
            
            x += 50;
        }
        
        x = 37;
        y += 50;
    }
}

void getCell(int x, int y) {
    if(x > 20 && x < 70) {
        col = 0;
    } else if(x < 120) {
        col = 1;
    } else if(x < 170) {
        col = 2;
    } else if(x < 220) {
        col = 3;
    } else if(x < 270) {
        col = 4;
    } else if(x < 320) {
        col = 5;
    } else if(x < 370) {
        col = 6;
    } else if(x < 420) {
        col = 7;
    } else if(x < 470) {
        col = 8;
    }
    
    if(y > 20 && y < 70) {
        row = 0;
    } else if(y < 120) {
        row = 1;
    } else if(y < 170) {
        row = 2;
    } else if(y < 220) {
        row = 3;
    } else if(y < 270) {
        row = 4;
    } else if(y < 320) {
        row = 5;
    } else if(y < 370) {
        row = 6;
    } else if(y < 420) {
        row = 7;
    } else if(y < 470) {
        row = 8;
    }
}

void grid() {
    background(255);
    fill(255);
    
    strokeWeight(9); // thick lines
    
    // outer box
    line(20, 20, 20, 470);
    line(20, 20, 470, 20);
    line(470, 470, 470, 20);
    line(470, 470, 20, 470);
    
    strokeWeight(3); // thin lines
    
    int x = 20;
    int y = 20;
    int w = 50;
    int h = 50;
    
    int[][] cells = new int[9][9];
    
    for(int i = 0; i < 9; i++) {
        for(int j = 0; j < 9; j++) {
            rect(x, y, w, h);
            x += 50;
        }
        
        x = 20;
        y += 50;
    }
    
    strokeWeight(7); // thick lines for 3x3 boxes
    
    // separating for 3x3 boxes
    line(20, 170, 470, 170); // top horizontal line
    line(20, 320, 470, 320); // bottom horizontal line
    line(170, 20, 170, 470); // left vertical line
    line(320, 20, 320, 470); // right vertical line
}

void instructions() {
    textSize(20);
    fill(0);
    
    for(int i = 0; i < 4; i++) {
        text("Instructions", 500, 50);
    }
    
    strokeWeight(3);
    line(500, 55, 602, 55);
    
    int ty1 = 85; // y position of the first line of the first paragraph
    int ty2 = 210; // y position of the first line of the second paragraph
    int ty3 = ty2 + 100; // y position of the first line of the third paragraph
    
    text("Using the numbers 1 through 9, fill in the grid", 500, ty1);
    text("so that each row, column, and 3x3 box ", 500, ty1 + 25);
    text("contains the each of the nine numbers only", 500, ty1 + 50);
    text("once.", 500, ty1 + 75);
    
    text("To select a cell, hover over a desired cell and ", 500, ty2);
    text("click on it. If you made a mistake, press", 500, ty2 + 25);
    text("0 to get rid of the number and clear the cell.", 500, ty2 + 50);
    
    text("If you want to change the puzzle, please", 500, ty3);
    text("press the r key.", 500, ty3 + 25);
}

void keyPressed() {
    if (key == '0') {
        num = 0;
        updateArray();
    } else if(key == '1') {
        num = 1;
        updateArray();
    } else if(key == '2') {
        num = 2;
        updateArray();
    } else if(key == '3') {
        num = 3;
        updateArray();
    } else if(key == '4') {
        num = 4;
        updateArray();
    } else if(key == '5') {
        num = 5;
        updateArray();
    } else if(key == '6') {
        num = 6;
        updateArray();
    } else if(key == '7') {
        num = 7;
        updateArray();
    } else if(key == '8') {
        num = 8;
        updateArray();
    } else if(key == '9') {
        num = 9;
        updateArray();
    } else if(key == 'r' || key == 'R') {
        setup();
    }
}

void mousePressed() {
    num = 0;
    getCell(mouseX, mouseY);
}

void puzzle1() {
    // www.websudoku.com – Evil Puzzle 8,655,032,120
    
    p[0][5] = 1;
    p[0][6] = 2;
    p[0][7] = 6;
    p[1][6] = 7;
    p[2][2] = 5;
    p[2][4] = 3;
    p[2][7] = 8;
    p[3][0] = 2;
    p[3][1] = 3;
    p[3][2] = 6;
    p[3][3] = 4;
    p[3][7] = 9;
    p[4][3] = 6;
    p[4][5] = 9;
    p[5][1] = 8;
    p[5][5] = 2;
    p[5][6] = 4;
    p[5][7] = 7;
    p[5][8] = 6;
    p[6][1] = 5;
    p[6][4] = 4;
    p[6][6] = 1;
    p[7][2] = 3;
    p[8][1] = 9;
    p[8][2] = 8;
    p[8][3] = 7;
    
    displayBoard();
}

void puzzle2() {
    // www.websudoku.com – Evil Puzzle 124
    
    p[0][5] = 3;
    p[0][8] = 6;
    p[1][3] = 7;
    p[1][8] = 9;
    p[2][2] = 4;
    p[2][4] = 6;
    p[2][7] = 8;
    p[2][8] = 2;
    p[3][4] = 9;
    p[3][7] = 7;
    p[4][1] = 1;
    p[4][2] = 6;
    p[4][6] = 8;
    p[4][7] = 5;
    p[5][1] = 2;
    p[5][4] = 3;
    p[6][0] = 2;
    p[6][1] = 8;
    p[6][4] = 5;
    p[6][6] = 6;
    p[7][0] = 3;
    p[7][5] = 8;
    p[8][0] = 7;
    p[8][3] = 1;
    
    displayBoard();
}

void puzzle3() {
    // www.websudoku.com – Medium Puzzle 8,136,223,360
    
    p[0][2] = 7;
    p[0][5] = 6;
    p[1][1] = 3;
    p[1][5] = 5;
    p[2][0] = 5;
    p[2][1] = 6;
    p[2][2] = 4;
    p[2][4] = 1;
    p[2][7] = 2;
    p[3][0] = 2;
    p[3][3] = 5;
    p[3][4] = 6;
    p[3][8] = 7;
    p[4][2] = 5;
    p[4][6] = 6;
    p[5][0] = 8;
    p[5][4] = 7;
    p[5][5] = 2;
    p[5][8] = 3;
    p[6][1] = 8;
    p[6][4] = 5;
    p[6][6] = 7;
    p[6][7] = 9;
    p[6][8] = 4;
    p[7][3] = 9;
    p[7][7] = 3;
    p[8][3] = 3;
    p[8][6] = 5;
    
    displayBoard();
}

void puzzle4() {
    // www.websudoku.com – Evil Puzzle 6,101,976,392
    
    p[0][4] = 6;
    p[0][5] = 1;
    p[0][8] = 5;
    p[1][0] = 8;
    p[1][4] = 5;
    p[2][6] = 8;
    p[2][7] = 4;
    p[2][8] = 7;
    p[3][1] = 5;
    p[3][6] = 9;
    p[4][0] = 6;
    p[4][3] = 9;
    p[4][5] = 4;
    p[4][8] = 8;
    p[5][2] = 3;
    p[5][7] = 1;
    p[6][0] = 7;
    p[6][1] = 9;
    p[6][2] = 1;
    p[7][4] = 8;
    p[7][8] = 2;
    p[8][0] = 5;
    p[8][3] = 7;
    p[8][4] = 3;
    
    displayBoard();
}

void puzzle5() {
    // www.websudoku.com – Medium Puzzle 2,981,621,299
    
    p[0][0] = 7;
    p[0][4] = 2;
    p[0][5] = 9;
    p[0][7] = 1;
    p[1][0] = 2;
    p[1][4] = 4;
    p[1][5] = 8;
    p[2][0] = 8;
    p[2][1] = 3;
    p[2][4] = 6;
    p[2][6] = 2;
    p[3][0] = 4;
    p[3][2] = 9;
    p[3][5] = 6;
    p[3][8] = 8;
    p[4][0] = 3;
    p[4][8] = 1;
    p[5][0] = 5;
    p[5][3] = 4;
    p[5][6] = 6;
    p[5][8] = 3;
    p[6][2] = 2;
    p[6][4] = 9;
    p[6][7] = 3;
    p[6][8] = 7;
    p[7][3] = 6;
    p[7][4] = 1;
    p[7][8] = 2;
    p[8][1] = 4;
    p[8][3] = 2;
    p[8][4] = 7;
    p[8][8] = 6;
    
    displayBoard();
}

void updateAndDisplay() {
    num = 0;
    
    grid();
    
    displayBoard();
    
    instructions();
    
    if(checkCompletion()) {
        completed();
    }
}

void updateArray() {
    p[row][col] = num;

    updateAndDisplay();
}

void setup() {
    size(900, 500);
    background(255, 255, 255);
    
    for(int i = 0; i < p.length; i++) {
        for(int j = 0; j < p[i].length; j++) {
            p[i][j] = 0;
        }
    }
    
    grid();
    choosePuzzle();
}
```

I really like solving puzzles, so I decided to create a Sudoku GUI so that people can solve puzzles on the computer. In that sense, it is personal because of my interests. I did not create any art, but I did make a grid and get puzzles from https://www.websudoku.com instead of making my own puzzles (since making sudoku puzzles can be very difficult).

The logic of what the GUI should do and how the user's answers can be checked to ensure that each row, column, and 3x3 box had the numbers 1 through 9 without being repeated was very intuitive for me because I have been solving sudoku puzzles for many years. Making the GUI with five different arrays for five different puzzles was also relatively easy. What I found difficult was making sure that the GUI would appear each time the user pressed a key was difficult when I only used one array for all five puzzles because each value in my array was being reset to 0 in the grid function (and I did not realize that I kept calling it and all the values were being reset to 0). Other than that, nothing was particularly challenging, but writing the code using the logic that I had was pretty time-consuming.

In the future, I would like to enhance this GUI and make it so that the user can see which numbers he/she typed in by making those numbers a different color than the numbers that were given. I would also like to make it so that the given numbers cannot be erased. I also want to include a check button so that users can check their answers as they go. That way, they will not have to get to the end if they make a mistake in the beginning.

For the stats, I used 4 global variables (one 2D integer array and three integer variables). I used 0 classes and 0 methods, but I did use 5 functions that returned a boolean type and 16 functions that returned nothing (void type) for a total of 21 functions. 3 functions were predefined for me by Processing (setup(), mousePressed(), and keyPressed()). I chose 5 puzzles from https://www.websudoku.com, and 1 of the puzzles was a Medium level puzzle. The other 4 puzzles were Evil level puzzles.
